## Zadanie 3

Zdefiniuj poniższą funkcję i sporządź jej wykres dla argumentów z danego przedziału:

$f(x)=x^2+5$

1. $x>-1$ oraz $x<1$

2. $x>-6$ oraz $x<6$

3. $x>0$ oraz $x<5$

Wspierając się dokumentacją Matplotlib. Dodaj do wykresu etykiety osi, tytuły wykresów i legendy.

Wskazówki: Deklaracja funkcji w Python.

W Pythonie bloki kodu (w tym również funkcje) są wyróżniane za pomocą wcięć. Do wizualizacji wyników w Pythonie używa się pakietu Matplotlib.

W pliku mają być zawarte: imię, nazwisko, opis wykonywanego ćwiczenia - formatowanie markdown w poszczególnych komórkach.

```python
import matplotlib.pyplot as plt
import numpy as np
def f_x(x):
  result = (x * x) + 5
  return result
```

Powyżej importuję bibliotekę *matplotlib* oraz *numpy*, a także deklaruje funkcję *f_x*, którą będziemy badać z wykorzystaniem zaimportowanej biblioteki na poszczególnych przedziałach.

# Ad. 1

```python
ad1 = np.linspace(-1, 1, 100)
ad1_r = []

for i in ad1:
  ad1_r.append(f_x(i))

plt.figure()
plt.plot(ad1, ad1_r, label="f(x)")
plt.title("Wykres dla x > -1 oraz x < 1")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

Korzystając z *np.linspace* utworzyłem sobie przestrzeń dla rozwiązania, a następnie korzystając z tej przestrzeni wykorzystałem wcześniej napisanej funkcji *f_x*. Rozwiązanie uzyskałem za pomocą pętli *for* po całej przestrzeni. Następnie korzystając z pakietu *matplotlib* wyrysowałem oraz opisałem rozwiązanie. Następne przykłady rozwiązuje analogicznie.

# Ad. 2

```python
ad2 = np.linspace(-6, 6, 100)
ad2_r = []

for i in ad2:
  ad2_r.append(f_x(i))

plt.figure()
plt.plot(ad2, ad2_r, label="f(x)")
plt.title("Wykres dla x > -6 oraz x < 6")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

# Ad. 3

```python
ad3 = np.linspace(0, 5, 100)
ad3_r = []

for i in ad3:
  ad3_r.append(f_x(i))

plt.figure()
plt.plot(ad3, ad3_r, label="f(x)")
plt.title("Wykres dla x > 0 oraz x < 5")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
